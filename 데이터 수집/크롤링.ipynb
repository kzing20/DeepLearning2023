```python
import os
import shutil
from bing_image_downloader.bing_image_downloader import downloader


directory = '/content/drive/MyDrive/딥러닝/team-project/크롤링 데이터/'
```

```python
query = '이병헌'

# 초기 디렉토리 만들기
if not os.path.isdir(directory+ '/' + query):
      os.makedirs(directory+ '/' + query)

downloader.download(query, limit=100,  output_dir='./', adult_filter_off=True, force_replace=False, timeout=60)
print(os.listdir(query))
for file_name in os.listdir(query):
  print(f'[Train Dataset] {file_name}')
  shutil.move(query + '/' + file_name, '/content/drive/MyDrive/딥러닝/team-project/크롤링 데이터/' + query + '/' + file_name)
```

```python
import os

path = '/content/drive/MyDrive/딥러닝/team-project/크롤링 데이터/'  # 폴더의 실제 경로로 수정해야 합니다.
folder_list = os.listdir(path)

# 폴더 내 파일 목록 가져오기
for folder_name in folder_list:
  file_list = os.listdir(path+folder_name+'/')

  # 파일 이름 변경
  for i, file_name in enumerate(file_list):
      # if file_name.endswith('.jpg'):  # .jpg 확장자인 파일만 변경
      new_file_name = f'{folder_name}_{i+1}.jpg'
      os.rename(os.path.join(path+folder_name+'/', file_name), os.path.join(path+folder_name+'/', new_file_name))
```
