```python
# 압축 파일 경로

# zip 파일들이 위치한 폴더 경로
zip_files_folder = "/content/drive/MyDrive/Mydata"

# 압축을 풀 폴더 경로
destination_folder ="./"
```

```python
import zipfile

for zip_file in os.listdir(zip_files_folder):
    if zip_file.endswith('.zip'):
        zip_path = os.path.join(zip_files_folder, zip_file)

        # 원래 파일 이름 추출
        original_file_name = os.path.splitext(zip_file)[0]

        # 원래 파일 이름으로 압축 해제
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(os.path.join(destination_folder, original_file_name))
            print(zip_path,'완료')
```

```python
import os
import shutil

# 압축 푼 폴더 목록을 불러올 폴더 경로
source_folder = "./"

# 압축 푼 폴더 목록 가져오기
# folders = [folder for folder in os.listdir(source_folder) if os.path.isdir(os.path.join(source_folder, folder))]
folders = [folder for folder in os.listdir(source_folder) if os.path.isdir(os.path.join(source_folder, folder)) and folder.isdigit()]
len(folders)
```

```python
output_folder = '/content/drive/MyDrive/dataset'
os.listdir(output_folder)
```

```python
for folder in folders:
  folder_path = os.path.join(source_folder, folder)
    # S001, S002, S005 폴더 목록
  subfolders = ['S001', 'S002', 'S005']

  for subfolder in subfolders:
      # 하위 디렉토리 경로
      subfolder_path = os.path.join(folder_path, subfolder)

      ssubfolders = ['L1', 'L2', 'L3', 'L4', 'L5']
      for ssubfolder in ssubfolders:
        # 하위 -하위 디렉토리 경로
        # 19062731/S001/L1/E01/
        ssfolder_path = os.path.join(subfolder_path,ssubfolder,'E01')
        # 대상 폴더 생성
        target_folder = os.path.join(output_folder,folder)
        os.makedirs(target_folder, exist_ok=True)

        # JPG 파일 필터링 및 복사
        files = os.listdir(ssfolder_path)
        for file in files:
            if file.endswith('.jpg') and file not in ['C1.jpg', 'C2.jpg', 'C12.jpg', 'C13.jpg']:
                source_file = os.path.join(ssfolder_path, file)
                output_file = os.path.join(target_folder, f"{subfolder}_{ssubfolder}_{file}")
                shutil.copy2(source_file, output_file)
                print(output_file, "완료")
```
