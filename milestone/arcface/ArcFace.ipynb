```python
from PIL import Image
def extract_face_feature(path):
  features = DeepFace.represent(path,detector_backend= 'mtcnn' ,model_name='ArcFace')
  img_embedding = features[0]['embedding']
  return img_embedding
```

```python
features = extract_face_feature(standard_path+standard_list[0])
features
```

```python
#기준 얼굴 feature vector 생성
standard_dic = []
standard_name=[]
for i in standard_list:
  path = '/content/drive/MyDrive/딥러닝/team-project/딥러닝/기준얼굴/'+i
  emd = extract_face_feature(path)
  standard_dic.append(emd)
  standard_name.append(i[:-4])
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm
import math
#유사도 정의 함수
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
#모든 파일 feature vector 생성
test_file = []
for i in image_list:
  path = '/content/drive/MyDrive/딥러닝/team-project/딥러닝/'+i
  emd = extract_face_feature(path)
  test_file.append(emd)
```

```python
len(test_file[0]) #512차원으로 embedding
```

```python
#cosine similarity
predict_label = []
for i in range(len(test_file)):
  test_similarity = []
  #각 기준파일과의 cosine similarity 계산 => 유사도가 가장 높은 이미지를 label로 사용
  for j in range(len(standard_dic)):
    sim = cos_sim(np.array(test_file[i]), np.array(standard_dic[j]))
    test_similarity.append(sim)
  max_index = test_similarity.index(max(test_similarity))
  max_name = standard_name[max_index]
  predict_label.append(max_name)
```

### 정확도 확인

```python
import unicodedata
num = 0
for i in range(100):
  #분리된 자음 모음 결합하는 코드
  label_predict = unicodedata.normalize('NFC',predict_label[i])
  label_actual = unicodedata.normalize('NFC',image_list_label[i])
  if label_predict == label_actual:
     num+=1
   
print("Accuracy: ", num/100)
```
